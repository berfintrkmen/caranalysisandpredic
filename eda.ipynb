```python
import pandas as pd
import numpy as np
import plotly as pl
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
data=pd.read_csv("cars_dataset.csv")
data.head()
```

```python
categoricals = list(data.select_dtypes('object').columns)
numericals = [col for col in data.columns if col not in categoricals]
print(categoricals)
print(numericals)
```

```python
sns.pairplot(data, corner=True)
```

```python
data[categoricals] = data[categoricals].astype('category')
data[categoricals].dtypes
```

```python
for col in data[categoricals]:
    print(f'We have {len(data[col].unique())} unique values in --{col}-- column: {data[col].unique()}', '\n')
```
